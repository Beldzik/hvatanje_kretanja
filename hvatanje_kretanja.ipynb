```python
# 163
# Kod ne radi iz nekog razloga
# Nije ni toliko bitno prikazuje neki tornado bukv mislim da i nije u nasem domenu projekta
from pylab import *
t,x,y,z = loadtxt('tornado.d', usecols=[0,1,2,3], unpack = True);
n = len(t);
dt = t[1] - t[0];
a = zeros((n, 3), float);
a[:, 0] = x;
a[:, 1] = y;
a[:, 2] = z;
v = zeros((n, 3), float);
r = zeros((n, 3), float);
r[0] = array([-80.0, 0.0, 0.0]);
v[0] = array([184.9, -18.49, 73.96]);
for i in range(0, n-1):
    v[i+1] = v[i] + a[i]*dt;
    r[i+1] = r[i] + 0.5*(v[i+1] + v[i])*dt;
    t[i+1] = t[i] + dt;
ddt = 1.0;
it = round(ddt/dt);
i = r_[i:it:n];
plot(r[i,0], r[i,1], 'o');
show();
```

```python
# 168-171
from pylab import *
import numpy as np
h = 10.0;
C = 30.0;
w0 = 3.0;
b = 5.0;
g = 9.8;
time = 20.0;
dt = 0.001;
n = int(ceil(time/dt));
t = zeros((n, 1), float);
r = zeros((n, 2), float);
v = zeros((n, 2), float);
a = zeros((n, 2), float);
t[0] = 0.0;
r[0,:] = array([0.0, h]);
v[1,:] = array([0.0, 0.0]);
for i in range(n-1):
    w = w0*(1.0 - exp(-r[i, 1]/b))*array([1, 0]);
    a[i,:] = -g*array([0, 1]) - C*norm(v[i,:] - w)*(v[i,:] - w);
    v[i+1,:] = v[i,:] + a[i,:]*dt;
    r[i+1,:] = r[i,:] + v[i,:]*dt;
    t[i+1] = t[i] + dt;
i = np.argwhere(r[:,1] > 0.0);
plot(r[i,0], r[i,1], '-k');
axis('equal'), xlabel('x [m]'); ylabel('y [m]');
show();
# Bilo bi pozeljno da ovaj graf pokazuje samo pozitivne vrednosti, odnosno grafik u oblasti y < 0 je suvisan
```

```python
# Motion Capture (180)
# Motion 1
from pylab import *
# g = 9.81;
t,ax,ay = loadtxt('motion1.d', usecols=[0,1,2], unpack = True);
n = len(t);
a = zeros((n, 2), float);
a[:, 0] = ax;
a[:, 1] = ay;
r = zeros((n, 2), float);
v = zeros((n, 2), float);
n_v = zeros((n, 1), float);
r[0,:] = array([0.0, 30.0]);
v[1,:] = array([0.0, 0.0]);
n_v[0] = 0;
dt = t[1] - t[0];
for i in range(n-1):
    v[i+1, :] = v[i,:] + a[i,:]*dt;
    r[i+1, :] = r[i,:] + v[i,:]*dt;
    n_v[i+1] = norm(v[i+1,:]);
xlabel('x [m]'); ylabel('y [m]'); title("Motion 1"); axis('equal');
plot(r[:, 0], r[:, 1], '-k');
show();
xlabel('t [s]'); ylabel('v [m/s]');
plot(t, n_v, '-k');
show();
```
